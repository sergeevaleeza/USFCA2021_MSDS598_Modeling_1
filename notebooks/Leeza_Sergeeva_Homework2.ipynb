```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
import random

import seaborn as sns
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python
df_taxis = sns.load_dataset('taxis')
df_taxis.head(10)
```

1. Suppose we want to use the variables distance and passengers to predict total which represents the total fare.

(a) Use linear regression to create a model prediction total from distance and passengers. Report the R2.

```python
model = ols(formula = 'total ~ distance + passengers', data=df_taxis)
res = model.fit()
res.summary()
```

R-squared value is 0.818, showing strong positive correlation between the total of the fair and number of passangers and distance they traveled.

(b) Incorporate another variable taxi color into the model and then report the R2.

```python
model = ols(formula = 'total ~ distance + passengers + color', data=df_taxis)
res = model.fit()
res.summary()
```

By adding a third variable of the color taxi to the previous model we get a new value for R-squared: 0.829. 

2. As in the fourth lecture, create a length of ride variable. Use the new feature you’ve engineered to further improve the model and report the R2.

```python
delta = pd.to_datetime(df_taxis['dropoff']) - pd.to_datetime(df_taxis['pickup'])

df_taxis['length_of_ride'] = delta / pd.Timedelta('60s')
# lengthh of ride is in minutes

model = ols(formula = 'total ~ distance + passengers + color + length_of_ride', data=df_taxis)
res = model.fit()
res.summary()
```

```python
print(f'This model has an R^2 of {res.rsquared}')
```

3. Create a validation set using ten percent of the data, and use the complement to create a training set. (Make sure there is no overlap in observations between the validation and training set).

```python
from sklearn.model_selection import train_test_split

# take random 10% of the taxi data for testing
train, test = train_test_split(df_taxis, test_size=0.1, shuffle = False)

# build model on training set
model = ols(formula = 'total ~ distance + passengers + color + length_of_ride', data=train)
res = model.fit()

# predict using the model
y_pred = res.predict(test)

# score the model
y = test['total']
y_mean = train['total'].mean()  # use the mean of the training set

TSS = sum((y_pred - y_mean)**2)
RSS = sum((y_pred - y)**2)
print(f'This model has an R^2 on the test set of {(TSS - RSS) / TSS}')
```
